# Predicting Late Shipments: A Machine Learning Approach

## Project Overview

Timely delivery is a **critical factor** in supply chain management, as late shipments can lead to **customer dissatisfaction, revenue loss, and increased operational costs**. In this project, I use **machine learning** to develop predictive models for a global sports and outdoor equipment retailer to **proactively identify high-risk shipments** before delays occur.

## Dataset Information
The data for this project is provided by DataCo and is publily available on Kaggle ([link](https://www.kaggle.com/datasets/shashwatwork/dataco-smart-supply-chain-for-big-data-analysis)). It contains approximately **180,000 transactions** that are shipped to **164 unique countries** over the span of **three years**. The dataset provides a **comprehensive view of supply chain operations**, including:
- **Order details:** Order IDs, order destination, and order dates.
- **Product information:** Product IDs, pricing, and sales.
- **Customer data:** Customer segment and locations.
- **Shipping records:** Scheduled vs. actual delivery times, shipping mode, and delivery status.

## Business Problem & Project Goals
The dataset reveals that 57% of completed shipments are late by at least one day, and more than 7% are delayed by three or more days. To improve on-time delivery rates, businesses need to identify high-risk shipments early and take proactive measures to mitigate delays. The goal of this project is to develop predictive models that can anticipate late deliveries, enabling supply chain teams to optimize logistics, allocate resources efficiently, and improve overall delivery performance.

#### **Objective**
I build two **Random Forest machine learning classifiers**:
1. **Late Order Model (Optimized for Accuracy)** – Predicts whether an order will be late (by at least 1 day).
2. **Very Late Order Model (Optimized for Recall)** – Predicts if an order will be **3+ days late**, prioritizing recall to flag high-risk shipments.

## Data Manegement

#### Import Required Libraries

```python
# Standard library
from pathlib import Path
import os
import sys

# Data handling
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Scikit-learn: Data Processing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import (
    MinMaxScaler, StandardScaler, RobustScaler, 
    OneHotEncoder, OrdinalEncoder
)

# Scikit-learn: Models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Scikit-learn: Evaluation Metrics
from sklearn.metrics import (
    accuracy_score, recall_score, precision_score, 
    f1_score, confusion_matrix, precision_recall_curve, 
    average_precision_score
)

# Scikit-learn: Hyperparameter Tuning
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
```

#### Set Project Directory Dynamically

```python
# Set BASE_DIR dynamically
if "ipykernel" in sys.modules:  # Checks if running in Jupyter Notebook
    from IPython import get_ipython
    notebook_path = Path(get_ipython().run_line_magic('pwd', '')).resolve()

    if notebook_path.name == "notebooks":
        BASE_DIR = notebook_path.parent
    else:
        BASE_DIR = notebook_path
else:
    # Fallback (if running in Spyder)
    BASE_DIR = Path().resolve()

# Define data directory
DATA_DIR = BASE_DIR / "data"
data_path = DATA_DIR / "DataCoSupplyChainDataset.csv"
```

#### Load & Explore Raw Data

```python
# Load dataset
df = pd.read_csv(data_path, encoding='latin1')

# Dataset dimensions
print(f"Dataset contains {df.shape[0]:,} rows and {df.shape[1]:,} columns.")

# Check column names and data types
print("\nColumn Names & Data Types:")
print(df.dtypes)

# Check missing values
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]  # Only show columns with missing values
if not missing_values.empty:
    print("\nMissing Values:")
    print(missing_values)
else:
    print("\nNo missing values found.")

```

#### Data Cleaning & Feature Engeneering

```python
# CLEAN DATA

# Remove incorrect product category names
df = df[df['Category Name'] != 'As Seen on  TV!']

# Remove incorrect customer states
df = df[~df['Customer State'].isin(['91732', '95758'])]

# Remove white spaces and extra spaces of customer streets
df['Customer Street'] = df['Customer Street'].str.strip().str.replace(r'\s+', ' ', regex=True)

# Keep only orders that have been successfully fulfilled or finalized
# "COMPLETE" - Orders that have been fully processed and delivered.
# "CLOSED" - Orders that have been finalized, possibly including returns or other adjustments.
df = df[df['Order Status'].isin(['COMPLETE', 'CLOSED'])]
```

```python
# CREATE NEW FEATURES

# Target variables
df['late'] = (df['Days for shipping (real)'] > df['Days for shipment (scheduled)']).astype(int)
df['very_late'] = (df['Days for shipping (real)'] > (df['Days for shipment (scheduled)'] + 2)).astype(int)

# Total order value
df['Order Value'] = df.groupby('Order Id')['Order Item Total'].transform('sum')

# Products per order
df['Unique Items per Order'] = df.groupby('Order Id')['Order Item Id'].transform('nunique')

# Number of units per order
df['Units per Order'] = df.groupby('Order Id')['Order Item Quantity'].transform('sum')

# Time features
df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'], format='%m/%d/%Y %H:%M')
df['year'] = df['order date (DateOrders)'].dt.year
df['month'] = df['order date (DateOrders)'].dt.month
df['day'] = df['order date (DateOrders)'].dt.day
df['hour'] = df['order date (DateOrders)'].dt.hour
df['minute'] = df['order date (DateOrders)'].dt.minute
df['day_of_week'] = df['order date (DateOrders)'].dt.weekday
```

```python
# DROP UNNECASSARY FEATURES
df = df.drop([
    'Days for shipping (real)',
    'Days for shipment (scheduled)',
    'Sales per customer',
    'Delivery Status',
    'Late_delivery_risk',
    'Category Name',
    'Customer Email',
    'Customer Fname',
    'Customer Id',
    'Customer Lname',
    'Customer Password',
    'Department Name',
    'Market',
    'Order Customer Id',
    'order date (DateOrders)',
    'Order Id',
    'Order Item Discount',
    'Order Item Id',
    'Order Item Profit Ratio',
    'Order Status',
    'Order Status',
    'Order Zipcode',
    'Product Category Id',
    'Product Description',
    'Product Image',
    'Product Name',
    'Product Status',
    'shipping date (DateOrders)',
    'Latitude',
    'Longitude'
], axis=1)
```

```python
# TRAIN-TEST SPLIT
X = df.drop(columns=['late', 'very_late'])
y_late = df['late']
y_very_late = df['very_late']
X_train, X_test, y_late_train, y_late_test = train_test_split(X, y_late, random_state=42)
y_very_late_train = y_very_late.loc[y_late_train.index]
y_very_late_test = y_very_late.loc[y_late_test.index]
```

```python
# FEATURE ENGINEERING: ENCODING AND SCALING

# Define feature groups
numerical_features = ['Benefit per order', 'Order Item Product Price', 'Order Item Quantity', 'Sales', 'Order Item Total',
                     'Product Price', 'year', 'month', 'day', 'hour', 'minute', 'Order Value', 'Unique Items per Order', 
                     'Order Item Discount Rate', 'Units per Order', 'Order Profit Per Order']
onehot_features = ['Type', 'Customer Segment', 'Shipping Mode']
label_features = ['Category Id', 'Customer City', 'Customer Country', 'Customer State', 'Customer Street', 'Customer Zipcode'
                 , 'Department Id', 'Order City', 'Order Country', 'Order Item Cardprod Id', 'Order Region', 'Order State'
                 , 'Product Card Id']

# Initialize scalers and encoders
scaler = RobustScaler() # Alternatively: MinMaxScaler()
onehot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
ordinal_encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)


# Apply robust scaler to numerical features (only fit using train data) and add indices and columns names from X_train anmd X_test
X_train_scaled = scaler.fit_transform(X_train[numerical_features])
X_test_scaled = scaler.transform(X_test[numerical_features])

X_train_scaled = pd.DataFrame(X_train_scaled, columns=numerical_features, index=X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=numerical_features, index=X_test.index)


# Process low cardinality categorical features using one-hot encoding
X_train_onehot = pd.DataFrame(
    onehot_encoder.fit_transform(X_train[onehot_features]),
    columns=onehot_encoder.get_feature_names_out(onehot_features),
    index=X_train.index
)

X_test_onehot = pd.DataFrame(
    onehot_encoder.transform(X_test[onehot_features]),
    columns=onehot_encoder.get_feature_names_out(onehot_features),
    index=X_test.index
)


# Process high cardinality categorical features using label encoding
X_train_label = pd.DataFrame(
    ordinal_encoder.fit_transform(X_train[label_features]),
    columns=label_features,
    index=X_train.index
)

X_test_label = pd.DataFrame(
    ordinal_encoder.transform(X_test[label_features]),
    columns=label_features,
    index=X_test.index
)


# Concatenate pre-preocessed features
X_train_final = pd.concat([X_train_scaled, X_train_onehot, X_train_label], axis=1)
X_test_final = pd.concat([X_test_scaled, X_test_onehot, X_test_label], axis=1)
```

```python
# DROP ADDITIONAL FEATURES
vars_to_drop = [
    'Customer City', # basically becomes a customer id
    'Customer Street', # same as above
    'Customer Zipcode', # same as above
    'minute', # irrelevant
    'hour', # irrelevant
    'Order Item Product Price', # perfectly correlated feature
    'Sales', # perfectly correlated feature
    'Order Item Cardprod Id', # perfectly correlated feature
    'Product Card Id', # perfectly correlated feature
    'Benefit per order'] # perfectly correlated feature


X_train_final = X_train_final.drop(vars_to_drop, axis=1)
X_test_final = X_test_final.drop(vars_to_drop, axis=1)
```

## Key insights
- Both classifiers perform very well in predicting late orders.

- The Late Order Model has an accuracy score of 92.1%. This means that the classifier is able to predict if an order will be late by one day or more in 92.1% of cases.

- The Very Late Order Model has a recall score of 97.3%. This means that the classifier is able to flag 97.3% of all orders that  will be at least 3 days late.

- The price for optimizing on recall and identifying almsot all very late orders is that the classifier also flags orders that will not be late. Only 72.3% of all orders flagged as very late will actually turn out to be very late. This is called the precision score of the classifier. If monitoring orders turns out to be very costly we can increase the precision of the model (i.e. reduce the number of flagged orders that are not late). The trade-off between identifying as many very late orders as possible (recall) and not flagging to many orders that will not be very late (precision) is captured by the precision-recall-curve. We can increase precision and lower recall by increasing the threshold for when orders are flagged. The current threshold is 0.3.

- The features that are most important to classify late orders are similar for both models. "Shipping Mode_Standard Class" is by far the most important with a feature importance score of 0.19 for the Late Order Model and 0.53 for the Very Late Model. Other important features include "Order Value", "Order City", "Order State", "day", and "Unites per Order".

- While both classifiers perform well there are many ways to potentially enhance performance. Here are three things one could consider going forward:
    1. We can use alternative learning methods to Random Forest to train the classifiers. Two vaiable alternatives are XGBoost or LightGBM. 
    2. We can work to prevent so called "leakage" of the models. It important that no features that we use to predict late orders include information on what happens after the point the customer makes the order. For instance, if customer gets a partial refund due to late delivery and this is reflected in the feature that captures order value, then we train our model on data it should not have access to. We need to ensure that no feature contains information that gives the classifer information on potential delivery success of an order.
    3. We can also try to gather more informaton about orders or generate new features. One example if this would be to gather coordinates on the locations where an order is being shipped from and delivered to, to calculate the shipping distance. We can then train the model on an extended set of features and see if any of the new features obtain a high importance score.
